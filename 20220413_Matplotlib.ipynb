# Figure.add_axes(size)
- **size**: [width_start, height_start, width_end, height_end]

```python
fig1 = plt.figure() # default figsize = [6, 4]
print(fig1)

a = fig1.add_axes([0, 0, 1, 1]) # Set size
print(type(a))

b = fig1.add_axes([0, 0, 0.5, 0.5])
c = fig1.add_axes([0, 0, 0.3, 0.3])
c.plot(xs, ys)
a.plot(xs, ys)

#d = fig1.add_axes([1, 0, 2, 1])
fig1.suptitle('Figure1', fontsize = 20)
# fig1.supxlabel('Figure1-Xlabel')
# fig1.supylabel('Figure1-Ylabel')
fig1.set_facecolor('green')
fig1.text(1.0, 1.0, 'Mytext', fontsize = 15)
```

- **Figure.add_subplot(nrow, ncol, index)**
  - **nrow**: number of row
  - **ncol**: number of column
  - **index**: current axe's position index

```python
fig2 = plt.figure()
a = fig2.add_subplot(2, 1, 1, title = 'Sin')
a.plot(xs, ys)

ys_2 = np.cos(xs)
b = fig2.add_subplot(2, 1, 2, title = 'Cos')
b.plot(xs, ys_2)
```

```python
fig2 = plt.figure()
a = fig2.add_subplot(1, 2, 1, title = 'Sin')
a.plot(xs, ys)

ys_2 = np.cos(xs)
b = fig2.add_subplot(1, 2, 2, title = 'Cos')
b.plot(xs, ys_2)
```

```python
fig2 = plt.figure()
a = fig2.add_subplot(2, 1, 1, title = 'Sin')
a.plot(xs, ys)

ys_2 = np.cos(xs)
b = fig2.add_subplot(2, 2, 3, title = 'Cos')
b.plot(xs, ys_2)

ys_3 = np.tan(xs)
c = fig2.add_subplot(2, 2, 4, title = 'Tan')
c.plot(xs, ys_3)
```

# Overlap is possible

# axe
- **axe.plot(x, y)**: plot값 (x, y)설정
- **axe.set_xlim**: x label 값 범위 설정
- **axe.set_ylim**: y label 값 범위 설정
- **axe.set_xticks( )**: x label에 특정 값만 tick mark

```python
f, axe = plt.subplots()
axe.plot(xs, ys)
a = axe.set_title('axe title')
b = axe.set_xlabel('X label')
c = axe.set_ylabel('Y label')

print(type(a), type(b), type(c))

axe.set_xlim(0, 10)
axe.set_ylim(-1.5, 1.5)

axe.set_xticks(np.linspace(0, 10, 3)) # numpy의 값에 해당하는 부분만 특정 label 표기 (tick marker)
```

- **axe.set_xmargin(ratio)**: ratio 만큼 x축으로 margin 추가
- **axe.set_ymargin(ratio)**: ratio 만큼 y축으로 margin 추가
- **axe.margins(ratio)**: x, y축 같이 ratio만큼 margin 추가

```python
f, axe = plt.subplots()
axe.plot(xs, ys)
a = axe.set_title('axe title')
b = axe.set_xlabel('X label')
c = axe.set_ylabel('Y label')

axe.set_xmargin(0.3)
axe.set_ymargin(0.3)
# axe.margins(0.3)   Same with previous two lines
```

# Matplotlib 활용 예제

```python
pd_data = pd.read_csv('/content/drive/MyDrive/인공지능 학습/Data/전국도서관표준데이터_utf8.csv', encoding = 'utf8', header = [0])
print(pd_data.shape)
print(pd_data.columns)
print(pd_data.head(5))
```

# Data pre-processing

```python
location_data = pd_data.loc[:, ['시도명', '자료수(도서)', '위도', '경도', '도서관유형']]
```

# Plotting

```python
plt.style.use('seaborn-white')

# Create figure outline
fig = plt.figure(tight_layout = True)
loc = fig.add_subplot(1, 2, 1)
lib_count = fig.add_subplot(2, 2, 2)
book_count = fig.add_subplot(2, 2, 4)

# Remove NaN (결측치)
location_data.dropna(inplace = True)
#print(location_data.loc[:, '위도'].hasnans, location_data.loc[:, '경도'].hasnans)
#location_data.plot.scatter('위도', '경도')

# Set values of library location
x1 = location_data['경도']
y1 = location_data['위도']
loc.scatter(x1, y1, s = 1)
loc.set_ylabel('lat', fontsize = 15)
loc.set_title('Library location', fontsize = 20)
loc.set_xlim(125.5, 130)
loc.set_xticks([])
loc.set_xlabel('long.', fontsize = 15)

#print(location_data.groupby('시도명').sum().shape)
# print(location_data.groupby('시도명').sum())
x2 = x3 = np.arange(0, 17)

# y2 = location_data.groupby('시도명').count().loc[:,'자료수(도서)']
#print(location_data['시도명'].value_counts())

y2 = location_data['시도명'].value_counts().sort_index() # Index의 순서대로 Sorting
lib_count.bar(x2, y2)

lib_count.set_title('Library count by area', fontsize = 20)
lib_count.set_xticks([])

y3 = location_data.groupby('시도명').sum().loc[:,'자료수(도서)']
book_count.bar(x3, y3)
book_count.set_title('Book count by area', fontsize = 20)
book_count.set_xticks([])
```

```python
b_filter = location_data['도서관유형'] == '공공도서관'
location_data = location_data[b_filter]
```

```python
print(location_data)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].hist(location_data['시도명'])
axes[0].set_xticks([])
axes[0].set_title('Library count by area', fontsize = 15)

x2 = location_data.groupby('시도명').sum()
print(len(x2['자료수(도서)']), x2['자료수(도서)'])
axes[1].bar([n for n in range(len(x2['자료수(도서)']))], x2['자료수(도서)'])
axes[1].set_xticks([])
axes[1].set_title('Book count by area', fontsize = 15)

plt.show()
```
